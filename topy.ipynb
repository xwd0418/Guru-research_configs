```python
def GreedyFindQuests(quests, target_gold):
    # returns the number of quests needed to reach target_gold
    # if it is impossible to reach target_gold, return None
    curr_gold = 0
    curr_exp = 0
    priority_q = [] # Max heap, stores all the quests that are doable. 
    sort quests from lowest_exp needed to highest_exp needed
    index = 0
    curr_quest = quests[index]
    result = 0
    while curr_gold < target_gold:
        if curr_quest.needed_exp > curr_exp:
            return None
        while  curr_quest.needed_exp <= curr_exp:
            priority_q.push(curr_quest)
            index += 1
            curr_quest = quests[index]
        curr_quest = priority_q.pop()
        result += 1
        curr_gold += curr_quest.gold_reward
        curr_exp += curr_quest.exp_reward    
   
    return result
```
